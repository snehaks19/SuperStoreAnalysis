#Bar plot between sub-category and category

```python
plt.figure(figsize=(16,8))
plt.bar("Sub-Category","Category",data=df)
plt.show()
```

#

#Pie plot of Sub-Category

```python
plt.figure(figsize=(6,5))
df["Sub-Category"].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

#Bar plot of sales vs profit

```python
df.groupby("Sub-Category")["Sales","Profit"].agg(["sum"]).plot.bar()
```

```python
print(df["State"].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=df["State"])
plt.xticks(rotation=90)
plt.show()

```

#Sales and profit region wise

```python
df.groupby('Region')['Profit',"Sales"].agg([sum]).plot.bar()
plt.title("Total profit and sales per region")
plt.rcParams['figure.figsize']=[10,0]
plt.show()
```

#Count plot of sub-categories

```python
print(df["Sub-Category"].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=df["Sub-Category"])
plt.xticks(rotation=90)
plt.show()
```

#Heat map of correlation among columns

```python
fig,axes=plt.subplots(1,1,figsize=(6,3))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

#Heat map of covariation between set of variables

```python
fig,axes=plt.subplots(1,1,figsize=(6,3))
sns.heatmap(df.cov(),annot=True)
plt.show()
```

#Line plot of discount vs profit

```python
plt.figure(figsize=(10,4))
sns.lineplot(x="Discount", y="Profit", data=df, color="r", label="Discount")
plt.show()

```
