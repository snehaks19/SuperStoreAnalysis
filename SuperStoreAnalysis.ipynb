#**Sales Analysis** 






#Total sales of each categories

```python
total_sales = df['Sales'].sum()
print("Total Sales: $", total_sales)
plt.figure(figsize=(8,8))
plt.bar("Category","Sales",data=df,color="brown")
plt.show()
```

Top selling category


```python
top_selling_category = df.groupby('Category')['Sales'].sum().nlargest(1)
print(top_selling_category)
```

#Total sales of each subcategory

```python
total_sales_Subcategory = df.groupby('Sub-Category')['Sales'].sum()
print("Total Sales: $", total_sales_Subcategory)
plt.figure(figsize=(16,8))
plt.bar("Sub-Category","Sales",data=df)
plt.xticks(rotation=90)
plt.show()
```

#top 10 cities with maximum sales

```python
top_10_city_sales = df.groupby('City')['Sales'].sum().nlargest(10)
print(top_10_cities_sales)
plt.figure(figsize=(16,5))
top_10_city_sales.plot(kind='bar')
plt.xticks(rotation=90)
plt.show()
```

#Top 10 states with maximum sales

```python
top_10_state_sales = df.groupby('State')['Sales'].sum().nlargest(10)
print(top_10_state_sales)
plt.figure(figsize=(16,8))
top_10_state_sales.plot(kind='bar',color="green")
plt.xticks(rotation=90)
plt.title("Top 10 states with maximum sales")
plt.show()
```

#Region sales

```python
region_sales = df.groupby('Region')['Sales'].sum().nlargest(10)
print(region_sales)
plt.figure(figsize=(4,2))
colors=['green','yellow','orange',"blue","Red"]
region_sales.plot(kind='bar',color=colors)
plt.xticks(rotation=90)
plt.show()
```

#Profit from Regions

```python
region_analysis = pd.DataFrame(df.groupby(['Region'])['Profit'].sum().reset_index())
region_analysis
plt.pie(region_analysis.Profit,
        labels = region_analysis.Region,
        startangle = 90, autopct = "%1.0f%%",
        explode = [0, 0, 0, 0.1])
plt.title("Most Profitable by Region")
plt.show()
```
